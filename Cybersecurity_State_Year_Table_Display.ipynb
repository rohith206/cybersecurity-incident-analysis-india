```python
import numpy as np
import pandas as pd

# Sample dataset (extended)
data = {
    'Year': [2019, 2020, 2021, 2021, 2022, 2022, 2023, 2023, 2023, 2023],
    'State': [
        'Tamil Nadu', 'Maharashtra', 'Karnataka', 'Delhi',
        'Maharashtra', 'Kerala', 'Karnataka', 'Telangana',
        'Gujarat', 'West Bengal'
    ],
    'Attack_Type': [
        'Phishing', 'Phishing', 'Ransomware', 'DDoS',
        'Malware', 'Phishing', 'DDoS', 'Ransomware',
        'Spyware', 'Phishing'
    ],
    'Target': [
        'Government', 'Individuals', 'Hospitals', 'Government',
        'Banking', 'Individuals', 'Government', 'Hospitals',
        'Corporates', 'Individuals'
    ],
    'Records_Lost': [8000, 5000, 30000, 10000, 25000, 7000, 12000, 27000, 15000, 6000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Take user input
state_input = input("Enter State name (e.g., Maharashtra): ").strip().title()
year_input = int(input("Enter Year (e.g., 2021): "))

# Filter based on input
filtered_df = df[(df['State'].str.title() == state_input) & (df['Year'] == year_input)]

# Display result as table
if not filtered_df.empty:
    from IPython.display import display
    display(filtered_df[['Attack_Type', 'Target', 'Records_Lost']])
else:
    print(f"\nNo data found for {state_input} in {year_input}.")
```
